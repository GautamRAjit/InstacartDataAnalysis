DROPPING THE MERGE COLUMN AND REDUCING THE DATA SIZE OF COLUMNS SO AS NOT TO RUN INTO MEMORY ISSUES

```python
ords_prods_merge.drop('_merge', axis=1, inplace=True)

#trying to decrease the datatype of the columns to avoid memory issues
ords_prods_merge['order_id'] = ords_prods_merge['order_id'].astype('int8')
ords_prods_merge['user_id'] = ords_prods_merge['user_id'].astype('int32')
ords_prods_merge['order_number'] = ords_prods_merge['order_number'].astype('int8')
ords_prods_merge['orders_day_of_week'] = ords_prods_merge['orders_day_of_week'].astype('int8')
ords_prods_merge['order_hour_of_day'] = ords_prods_merge['order_hour_of_day'].astype('int8')
ords_prods_merge['days_since_prior_order'] = ords_prods_merge['days_since_prior_order'].astype('int8')
ords_prods_merge['aisle_id'] = ords_prods_merge['aisle_id'].astype('int8')
ords_prods_merge['department_id'] = ords_prods_merge['department_id'].astype('int8')
ords_prods_merge['product_id'] = ords_prods_merge['product_id'].astype('int8')
ords_prods_merge['prices'] = ords_prods_merge['prices'].astype('float16')
```

Classifying the products as low, mid range and high range products

```python
#price_label function
res = []

for value in ords_prods_merge["prices"]:
  if value <= 5:
    res.append("Low range product")
  elif value > 5 and value <= 15:
    res.append("Mid range product")
  else:
    res.append("High range product")
```

```python
#creating the price_range column
ords_prods_merge['price_range'] = res
```

Q.Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result.append("Busiest days")
  elif value == 3 or value == 4:
    result.append("Slowest days")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
ords_prods_merge['busiest_days'] = result
```

Q.The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

Since the manager wants the 24 hours to be divided into 3 segments based on how many orders are received during the particular hour, we are dividing the hours into 3 groups of 8.
10,11,14,15,13,12,16,9 - Most orders
17,8,1,19,20,7,21,22 - Average orders
23,6,0,1,5,2,4,3 - Fewest orders

```python
results_2 = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value in [10,11,14,15,13,12,16,9]:
    results_2.append("Most orders")
  elif value in [17,8,1,19,20,7,21,22]:
    results_2.append("Average orders")
  else:
    results_2.append("Fewest orders")
```

```python
#creating the busiest period of day column
ords_prods_merge['busiest_period_of_day'] = results_2
```
