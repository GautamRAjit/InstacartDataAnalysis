TASK STEPS

Consistency checks for df_ords

```python
df_ords.describe()
```

```python
#checking for mix types
for col in df_ords.columns.tolist():
  weird = (df_ords[[col]].applymap(type) != df_ords[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_ords[weird]) > 0:
    print (col)
```

```python
#missing values in df_ords
df_ords.isnull().sum()
```

There are 206209 records with the field 'days_since_prior_order' missing! We need to subset the dataframe before proceeding

```python
#subsetting records where product name is not NULL
df_ords_clean = df_ords[df_ords['days_since_prior_order'].isnull() == False]
df_ords_clean.shape
```

```python
#checking for duplicates in df_ords
df_ords_dups = df_ords_clean[df_ords_clean.duplicated()]
df_ords_dups
```

There are no duplicates in the cleaned df_ords dataframe!

Consistency checks for df_prods

```python
for col in df_prods.columns.tolist():
  weird = (df_prods[[col]].applymap(type) != df_prods[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_prods[weird]) > 0:
    print (col)
```

```python
#cleaning df_prods where only records with product names are considered
df_prods = df_prods[df_prods['product_name'].isnull() == False]
df_prods.shape

for col in df_prods.columns.tolist():
  weird = (df_prods[[col]].applymap(type) != df_prods[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_prods[weird]) > 0:
    print (col)
```

```python
#df_prods.describe()
df2 = df_prods[df_prods['prices'] < 14000]
df2.describe()
```

```python
Two prices were found to be exceptionally high - 14900 and 99999. These were considered as outliers and removed from further analysis
```

```python
#looking for duplicates in the df_prods dataframe
df_dups = df2[df2.duplicated()]
df_dups
```

```python
#getting rid of duplicates in df_prods
df_prods_clean_no_dups = df2.drop_duplicates()
df_prods_clean_no_dups
```
